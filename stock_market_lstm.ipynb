```python
df["Tomorrow"]=df["Close"].shift(-1)
```

```python
df.head()
```

```python
df["Target"]=(df["Tomorrow"] > df["Close"]).astype(int)
```

```python
df.head()
```

```python
df = df.rename(columns={'Total Trade Quantity': 'Volume'})
```

```python
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(n_estimators=100, min_samples_split=100,random_state=1)

train=df.iloc[:-100]
test=df.iloc[-100:]

predictors=["Close","Volume","Open","High","Low"]
model.fit(train[predictors],train["Target"])
```

```python
from sklearn.metrics import precision_score
preds=model.predict(test[predictors])
```

```python
preds=pd.Series(preds,index=test.index)
```

```python
precision_score(test["Target"],preds)
```

```python
combined=pd.concat([test["Target"],preds],axis=1)
```

```python
combined.plot()
```

```python
def predict(train,test,predictors,model):
    model.fit(train[predictors],train["Target"])
    preds=model.predict(test[predictors])
    preds=pd.Series(preds,index=test.index,name="Predictions")
    combined=pd.concat([test["Target"],preds],axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=250, step=250):
    all_predictions = []
    
    # Iterate over the data with the specified start and step
    for i in range(start, len(data), step):
        train = data[:i]
        test = data[i:i+step]
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    # Check if there are valid predictions to concatenate
    if len(all_predictions) > 0:
        return pd.concat(all_predictions)
    else:
        return None
```

```python
predictions=backtest(df,model,predictors)
```

```python
if predictions is not None:
    predictions["Predictions"].value_counts()
else:
    print("No valid predictions available")
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = df.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    df[ratio_column] = df["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    df[trend_column] = df.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors+= [ratio_column, trend_column]
```

```python
df = df.dropna(subset=df.columns[df.columns != "Tomorrow"])
```

```python
df
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=5, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >=.5] = 1
    preds[preds <.5] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(df, model, new_predictors)
```
